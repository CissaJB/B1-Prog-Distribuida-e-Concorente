Inicialmente, ao fazer a questão, utilizei o Google Colab. No entanto, ao trabalhar com threads, encontrei problemas, pois, ao meu ver, elas não estavam funcionando corretamente. Pesquisando, descobri que isso ocorria devido ao GIL (Global Interpreter Lock), que impede que múltiplas threads sejam executadas simultaneamente. Como resultado, mesmo utilizando threads, os dados eram exibidos um de cada vez: primeiro, todos os formandos de um determinado curso, depois os de outro, e assim sucessivamente.

A solução que encontrei na internet para esse problema foi utilizar multiprocessing em vez de threads, o que resultou no comportamento esperado.

Ao testar o mesmo código no VSCode, as threads funcionaram corretamente. Isso me levou a pesquisar novamente o motivo dessa diferença e descobri que se deve ao fato de essa ser uma tarefa I/O-bound (leitura de arquivos). Como o Python libera temporariamente o GIL durante operações de entrada e saída, outras threads podem ser executadas enquanto aguardam a conclusão da leitura.

Portanto, as threads funcionam normalmente no VSCode ao rodar um arquivo com extensão .py, mas apresentam problemas ao serem usadas no ambiente Jupyter Notebook, que é utilizado tanto no Google Colab quanto no VSCode ao rodar arquivos com extensão .ipynb.
